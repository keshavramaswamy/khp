```python
def create_bool_feature(df,col,val,val0,val1):
    vals = list()
    for i, row in df.iterrows():
        if ((type(row[col]) is float) | (type(row[col]) is int)):
            if np.isnan(row[col]):
                vals.append(row[col])
            elif row[col] == val:
                vals.append(val0)
            else:
                vals.append(val1)
        else:
            if pd.isnull(row[col]):
                vals.append(row[col])
            elif row[col] == val:
                vals.append(val0)
            else:
                vals.append(val1)
    df[col+'_bool'] = vals
    return df
```
